```python
# Plot the marginal distribution at M_chirp_source = 1.2188 M_sun
mass2 = np.array(mass2)
tides2 = np.array(tides2)
prob2 = np.array(prob2)
# Normalise probability 
prob_total2 = sum(prob2)
print(prob_total2)
prob2 = np.array(prob2)/prob_total2
#print(tides[0:1000])
# Create a grid for the points
points = np.column_stack((mass2, tides2))
mass_sort2=np.sort(np.unique(mass2))
tides_sort2=np.sort(np.unique(tides2))

f_interp = CloughTocher2DInterpolator(points,prob2)
chirp_mass = 1.2188
y0_tmp = np.tile(chirp_mass,len(tides_sort2))
#np.interp(y0,mass,min_lam)
yy = f_interp(y0_tmp,tides_sort2)
print(max(yy))
plt.plot(tides_sort2,yy)
plt.axvline(340,label=r'$\tilde{\Lambda} = 340$',color='blue')
plt.axvline(313,label=r'$\tilde{\Lambda}_{\rm SLY} = 313$',color='teal')
plt.ylabel('Density')
plt.margins(x=0,y=0)
plt.ylim(0,8e-5)
plt.legend()
plt.xlabel(r'$\Lambda$')
plt.savefig('1.4+1.4_marg_lambda.png',dpi=500,bbox_inches='tight')
```

```python
# Make new pdf's for NSBH case
import glob as glob
import os 

directory ="/Users/smag0001/Downloads/BHNS/"
dat_files = glob.glob(os.path.join(directory, "*.dat"))
print(dat_files)
```

```python
def plot_NSBH(mass,tides,prob,file):

    
    plt.scatter(tides,mass,c=np.log10(prob),cmap='Purples',alpha=1.0)
    #plt.axhspan(1.186, 1.192, color='k', alpha=0.4)
    #plt.annotate(r'GW170817', (800,1.212), fontsize=20)

    # GW190425
    # plt.axhline(1.44, c='k', ls='-.')
    plt.axhspan(1.42, 1.46, color='k', alpha=0.4)
    plt.annotate(r'GW190425', (800,1.48), fontsize=20)
    #GW190814,
    plt.axhspan(6.09-0.06, 6.09+0.06, color='k', alpha=0.4)
    plt.annotate(r'GW190814', (800,6.09), fontsize=20)
    #GW230529
    plt.axhspan(1.94-0.04, 1.94+0.04, color='k', alpha=0.4)
    plt.annotate(r'GW230529', (800,2.0), fontsize=20)

    # Les entfants terrible
    # 105 is TOO LARGE
    #GW200105,
 
    #GW200115
    plt.axhspan(2.58-0.007, 2.58+0.006, color='k', alpha=0.4)
    plt.annotate(r'GW200115', (800,2.6), fontsize=20)

    plt.colorbar(label=r'$\rm \log_{10}[\pi(\mathcal{M},\,\tilde{\Lambda})] $')
    plt.xlim(0, 1000)
    plt.ylim(min(mass), max(mass))
    #plt.ylim(0.8030150753768844,1.996984924623116)
    plt.ylabel(r'$\mathcal{M}\,\left[M_\odot\right]$')
    plt.xlabel(r'$\tilde{\Lambda}$')
    fig_filepath = '/Users/smag0001/Downloads/BHNS/plots/' +os.path.basename(file).replace('.dat', '.png') 
    plt.savefig(fig_filepath,dpi=500,bbox_inches='tight')
    plt.clf()
    
    
```

```python
for file in dat_files:
    # Create data frame from file 
    # Same thing for BNS wo GW constraints
    df_file = pd.read_csv(file,delim_whitespace=True)
    #df = df.sort_values(by=[‘mass’])
    print(df_file)

    mass = df_file['mass'].tolist()

    tides= df_file['tides'].tolist()
    counts = df_file['counts'].tolist()
    prob = np.array(counts)/sum(counts)
    tides = np.array(tides)
    mass = np.array(mass)
    
    # Check that prob is normalised 
    print(np.sum(prob))

    # Now save the pdf's to a csv 
    # Create dict 
    NSBH_pdf_dict = {'tides':tides,'mass':mass,'prob':prob}
    # Create df from dict 
    NSBH_df = pd.DataFrame(NSBH_pdf_dict)
    filepath_NSBH = '/Users/smag0001/Downloads/BHNS/prob_MCL/'+os.path.basename(file)
    print(filepath_NSBH)
    # Create csv from df
    NSBH_df.to_string(filepath_NSBH, index=False)

    # Make a plot and save to png in an output directory 
    plot_NSBH(mass,tides,prob,file)

    
    
    
```
